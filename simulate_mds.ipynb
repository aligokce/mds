# Classical MDS

```python
import numpy as np
```

```python
import mds
```

```python
p = 2 # Dimensions
n = 6 # Microphone count
roomSize = 5 # Room size (in p dimensions)
```

```python
# Randomly initialize microphone positions
np.random.seed(42)
micPos = roomSize * np.random.rand(n, p)
```

```python
micPos_est = mds.classical_from_points(micPos, p)
```

```python
mds.rms(micPos_est, micPos)
```

## Plots

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,4))
mds.plot_2d(micPos, ax=axs[0])
mds.plot_2d(micPos_est, remove_min=True, ax=axs[1])
plt.xlim([0, roomSize])
plt.ylim([0, roomSize])
plt.show()
```

# Basis-point MDS

```python
import numpy as np
```

```python
import mds
```

```python
p = 3 # Dimensions
n = 7 # Microphone count
roomSize = 5 # Room size (in p dimensions)
```

```python
# Randomly initialize microphone positions
np.random.seed(42)
micPos = roomSize * np.random.rand(n, p)
```

```python
micPos_est, chosen_indices = mds.basis_point_from_points(micPos, 3)
```

```python
mds.rms(micPos_est, micPos)
```

## Plots

```python
%matplotlib widget
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1, 2, 1, projection='3d')
mds.plot_3d(micPos, chosen_indices, ax)

ax = fig.add_subplot(1, 2, 2, projection='3d')
mds.plot_3d(micPos_est, chosen_indices, ax)

plt.show()
```
